```python
new_dataFrame=new_dataFrame.drop(['transmission','model'],axis=1)
```

```python
new_Array=new_dataFrame.iloc[:].values 
```

```python
new_dataFrame['fuelType'].unique() 
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
```

```python
new_Array[:,3]=label_encoder.fit_transform(new_Array[:,3])
```

```python
new_Array
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct=ColumnTransformer(transformers=[('encode',OneHotEncoder(),[3])])
```

```python
numerical_array=ct.fit_transform(new_Array)
```

```python
numerical_frame=pd.DataFrame(numerical_array)
```

```python
numerical_frame.isna().sum()
```

```python
new_dataFrame=new_dataFrame.drop('fuelType',axis=1)
```

```python
b=pd.concat([new_dataFrame,numerical_frame],axis=1)
```

```python
myDataFrame=b.dropna(axis=0)
```

```python
y=myDataFrame['price'].values
```

```python
x=myDataFrame.drop('price',axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=17)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model=Sequential()
model.add(Dense(16,activation='relu'))

model.add(Dense(16,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(16,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
earlystopping=EarlyStopping(monitor='val_loss',mode='min',patience=20)
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),batch_size=200,epochs=500,callbacks=[earlystopping])
```

```python
x_train.shape
```

```python
lossData=pd.DataFrame(model.history.history)
```

```python
lossData.head()
```

```python
lossData.plot()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
predict=model.predict(x_test)
```

```python
mean_absolute_error(y_test,predict)
```

```python
plt.scatter(y_test,predict)
plt.plot(y_test,y_test,'r-*')
```
